- 归一化公式：$$normalized=\frac{value−min​}{max−min}×2−1$$
- 
绝对值之总和为1：将每个归一化值除以所有归一化值之和。

```python
# 定义数据
values = [-0.21, 0.41, -0.04, -0.25, 0.09]
# 第一步：归一化到[-1, 1]
min_val = min(values)
max_val = max(values)
normalized_values = [(val - min_val) / (max_val - min_val) * 2 - 1 for val in values]
# 第二步：归一化总和为1
sum_abs_values = sum([abs(val) for val in normalized_values])
final_normalized_values = [round(val / sum_abs_values,2) for val in normalized_values]
final_normalized_values
```
